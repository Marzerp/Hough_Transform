```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('resultados.csv')
```

```python
df
```

```python
promedios = df.groupby('procesos')['tiempo'].mean()
plt.figure(figsize=(8,5))
ax = sns.barplot(x=promedios.index, y=promedios.values, palette="viridis")
plt.xlabel('Proceso')
plt.ylabel('Tiempo promedio (s)')
plt.title('Tiempo promedio por proceso')
plt.show()
```

```python
# Gráfico de líneas del tiempo de cada proceso
plt.figure(figsize=(10,6))
for proceso in range(1,5):
    datos = df[df['procesos'] == proceso]
    plt.plot(datos['ejecucion'], datos['tiempo'], marker='o', label=f'Proceso {proceso}')
plt.xlabel('Ejecución')
plt.ylabel('Tiempo (s)')
plt.title('Tiempo por ejecución para cada proceso')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfica del promedio de tiempo por proceso con desviación estándar
promedios = df.groupby('procesos')['tiempo'].mean()
desviaciones = df.groupby('procesos')['tiempo'].std()
plt.figure(figsize=(8,5))
plt.errorbar(promedios.index, promedios.values, yerr=desviaciones.values, fmt='o-', color='b', capsize=5, label='Promedio ± Desv. Est.')
plt.xlabel('Procesadores')
plt.ylabel('Tiempo promedio (s)')
plt.title('Tendencia del tiempo promedio por proceso con desviación estándar')
plt.grid(True)
plt.xticks(promedios.index)
plt.legend()
plt.show()
```
